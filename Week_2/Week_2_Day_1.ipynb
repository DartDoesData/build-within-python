# `range()` function

To loop through code a certain number of times, we can use the range() function.

It takes up to three arguments: `start`, `stop`, and `step`.

- `start` is optional and is used to specify the starting value of the sequence (0 by default).
- `stop` is required and is used to specify the end value of the sequence (up to, but not including).
- `step` is optional and is used to specify how to iterate through the sequence (1 by default).

## `while` loops

A while loop in Python is a type of loop that allows you to repeatedly execute a block of code as long as a certain condition is true.

```python
# Set account balance
balance = 500

# Withdraw $50 at a time
while balance > 0:
    balance = balance - 50
    print(f'You just withdrew $50. Your new balance is {balance}.')

# Outside of the while loop
print(f'Your new balance is {balance}, which is no longer sufficient to withdraw $50.')
```
