## `for` loop

A for loop in Python allows you to iterate over a sequence of items (such as a list, tuple, string, or range) and perform a specific operation for each.

```python
character_name = 'Sabrina Spellman'

# Loop through a string
for i in character_name:
    print(i)
```

## `while` loops

A while loop in Python is a type of loop that allows you to repeatedly execute a block of code as long as a certain condition is true.

```python
# Set account balance
balance = 500

# Withdraw $50 at a time
while balance > 0:
    balance = balance - 50
    print(f'You just withdrew $50. Your new balance is {balance}.')

# Outside of the while loop
print(f'Your new balance is {balance}, which is no longer sufficient to withdraw $50.')
```

## Lists and Tuples

Lists and tuples allow you to store and access a set of values in a single variable. They can store contain any type of data - integers, floating-point numbers, strings, and even other lists.

Tuples are similar, except that tuples are designed to be immutable. In other words, you canâ€™t change the values in them. This is helpful for fixed data sets, such as geographic coordinates, or other things that donâ€™t need to be altered in your program.

image.png

```python
# Lists are created using square brackets

# The code snippet below will create a list called `names` containing five items.
names = ['Sabrina', 'Ambrose', 'Hilda', 'Zelda', 'Harvey']

# The code snippet below will print the `names` list to the Python console.
print(names)

# List items can be of any data type
example_list = ['Sabrina', 16, ['Prudence', 'Dorcas', 'Agatha'], True, 21, False]
print(example_list)

# Empty list
test_list = []
```

## What is Mutability?

Strings are â€œimmutableâ€, which means that we cannot change the contents of an existing string. Instead, we need to create a new string in order to change its value.

However, lists are â€œmutableâ€ which means that we can change an element of a list using the index operator. You access list elements through the index, which starts at 0.


```python
character = 'Sabrina'

print(character[4])

character[4] = 'n' # ERROR
```

```python
# Declaring a names list
names = ['Sabrina', 'Ambrose', 'Hilda', 'Zelda', 'Harvey']

# Change Harvey to Nick
names[4] = 'Nick'

#Print the new list
names
```

## Letâ€™s explore a few functions to modify lists:

- `append`: Weâ€™ll use the append function to append an item to the end of the list
- `insert`: Use the insert function to add an item at a specific position
- `remove`: The remove function finds an removes an item from a list
- `sort`: Sorts a list in ascending order.
  - Use `reverse=True` to sort in descending order
- `len`: Returns the number of items in a list (this doesnâ€™t change the list)


```python
# Create a new list called `names`
names = ['Sabrina', 'Ambrose', 'Hilda', 'Zelda', 'Harvey']
print(names)

# Appending items to the end of the list
names.append('Faustus')
print(names)

# Appending items to a specific position in the list
names.insert(1, 'Agatha')
print(names)

# Removing an item from a list
names.remove('Hilda')
print(names)

# Sort the list in ascending order
names.sort()
print(names)

# Sort the list in descending order
names.sort(reverse=True)
print(names)

# Get the number of items in a list
len(names)
```

## Tuples

Now letâ€™s move on to tuples. Tuples are very similar to lists, except their values canâ€™t changed, which makes them more efficient, especially for large file processing. Also, as opposed to lists, tuples are enclosed in parentheses ( ), and separated by commas. Lists are enclosed in square brackets.

```python
# Tuples are enclosed in parenthesis
customers = ('Sabrina', 'Ambrose', 'Hilda', 'Zelda', 'Harvey')

# Define variables in the form of a tuple
(lat, lng) = (38.06, -77.13)
print(lng)
```

# Dictionaries

## Components of a Python Dictionary

- `key`: used to identify a specific value in a dictionary. It serves as an index to access the corresponding value. Keys must be unique in a dictionary and can be of any immutable data type such as strings, numbers, and tuples.

- `value`: the data that is associated with a key in a dictionary. Values can be of any data type and can be repeated across different keys in a dictionary.

- `item` (key-value pair): a combination of a key and a value in a dictionary. It represents a single element in a dictionary.


## Creating a Python Dictionary

While there are many ways to create a list of dictionaries, we will focus on these two:
1. List of dictionaries: One Python list that contains the contents of several dictionaries
  - For example, a list of customers (customer dictionary) where each dictionary contains corresponding data, such as last name, first name, and customer ID)

2. Dictionary of lists: Several python lists combined into one dictionary
  - For example, three separate lists â€“ one that contains customer last names, another that contains customer first name, and a third with customer ID can be used and combined into one dictionary where the key describes the contents of the list.


```python
# Create a new dictionary
customer = {
  'last_name': 'Spellman',
  'first_name': 'Sabrina',
}

# Return their first name
customer['first_name']
```

```python
# Create a new dictionary
customer['birth_date'] = '10/31/2000'
customer['superpower'] = 'Levitation'
customer['superpowers'] = ['Levitation', 'Conjuring fire', 'Telepathy']

# Preview the dictionary
print(customer)
```

```python
# Pretty print
from pprint import pprint
```

```python
pprint(customer)
```

```python
pprint(customer, sort_dicts=False)
```

```python
# Removing items from a dictionary
del customer['birth_date']

# Preview the dictionary
print(customer)
```

```python
pprint(customer, sort_dicts=False)
```

```python
# Check to see if a value exists
# Does the value Sabrina exist in the dictionary?
'superpowers' in customer
```

```python
if('superpowers' in customer):
    print('This customer has superpowers!')
else:
    print('No superpowers!')
```

```python
# List the keys in a dictionary
print(customer.keys())

# List the values in a dictionary
print(customer.values())
```

```python
# Lists of dictionaries from individual dictionaries
sabrina_dict = {'last_name': 'Spellman',
                'first_name': 'Sabrina',
                'birth_date': '10/31/2000'}

nick_dict = {'last_name': 'Scratch',
                'first_name': 'Nicholas'}

harvey_dict = {'last_name': 'Kinkle',
                'first_name': 'Harvey'}

characters = [sabrina_dict, nick_dict, harvey_dict]

print(characters)
```

```python
# Create a list of dictionaries within one list
characters = [
    { 'last_name': 'Spellman',
      'first_name': 'Sabrina',
      'birth_date': '10/31/2000'},
    { 'last_name': 'Scratch',
      'first_name': 'Nicholas'},
    { 'last_name': 'Kinkle',
      'first_name': 'Harvey'
    }
]

print(characters)
```
