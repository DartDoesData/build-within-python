# Python Functions

Now, let’s get into Python functions. Functions are like mini-programs that perform a task or set of tasks. They allow you to break up your code into smaller, more manageable pieces.

Functions are helpful because you can avoid writing repeat code, make your programs more clean and flexible. You can also reuse functions and their logic in other parts of your program.

A simple example of a function is one that will combine first and last name. For example, if I’ve got Sabrina and Spellman stored separately in a database and I want to print the full name, Sabrina Spellman, you can use a function to combine those whenever you need to print the full name.

In Python, you could call this function something like `get_full_name()`


# Working with Functions

- Defining a function: Use the `def` keyword to write functions.
- Calling a function: You call a function by using the function name, followed by parenthesis.
- Passing arguments:
  - Functions can accept arguments, which are variables that help instruct the function work.
  - You call a function by using the function name along with its arguments in parenthesis.

- Returning values: Functions can also send back data with the return keyword, which is referred to as returning values.
Code blocks: Code is contains within code blocks, much like conditional statements and loops.

```python
accounts = [{'acct_no': '7779311', 'full_name': 'Sabrina Spellman', 'acct_bal': 19000},
           {'acct_no': '7789422', 'full_name': 'Ambrose Spellman', 'acct_bal': 21000},
           {'acct_no': '7829366', 'full_name': 'Hilda Spellman', 'acct_bal': 25000},
           {'acct_no': '7839477', 'full_name': 'Zelda Spellman', 'acct_bal': 32000}]

acct_df = pd.DataFrame(accounts)
acct_df
```

```python
# select more than one column: passing in a list of column names
client_bal = acct_df[['full_name','acct_bal']]

# note that this returns a DataFrame
client_bal
```

```python
# Statistical summary
client_bal.describe()
```

```python
# record count
print('# of accounts')
print(client_bal['acct_bal'].count())

print('Lowest account balance')
# min
print(client_bal['acct_bal'].min())

# max
print('Highest account balance')
print(client_bal['acct_bal'].max())

# mean
print('Average account balance')
print(client_bal['acct_bal'].mean())

# sum
print('Total account balances')
print(client_bal['acct_bal'].sum())
```
